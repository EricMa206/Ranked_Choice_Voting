## Instant runoff voting (IRV) in the 2009 Burlington, VT mayoral election

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [8,10]
plt.style = "fivethirtyeight"
```

```python
second_round = {"Candidate":["Bob Kiss", "Kurt Wright", "Andy Montroll"], 
                "Party": ["Progressive", "Republican", "Democrat"],
                "Votes": [2981, 3294, 2554],
                "%": ["33.2%", "36.7%", "28.4%"]}
```

```python
third_round = {"Candidate":["Bob Kiss", "Kurt Wright", "Andy Montroll"], 
                "Party": ["Progressive", "Republican", "Democrat"],
                "plus": [1332, 767, -2554],
                "Votes": [4313, 4061, 0],
                "%": ["48.0%", "45.2%", "28.4%"]}
```

```python
party = ["Progressive", "Republican", "Democrat"]
round2 = [2981, 3294, 0]
round3plus = [1332, 767, 0]
```

```python
round2_df = pd.DataFrame(second_round)
```

```python
round3_df = pd.DataFrame(third_round)
```

```python
plt.bar(round2_df.Party, round2_df.Votes, color=["Green", "Red", "Blue"])
plt.ylim(top=4500)
plt.title("Second Round Voting Results")
```

```python
plt.bar(party, round2, label='Round 2 Votes', color=["Green", "Red", "Blue"])
plt.bar(party, round3plus, label='Final Votes', color='Blue', bottom=round2)
plt.ylim(top=4500)
plt.title("Final Round Voting Results")
```
